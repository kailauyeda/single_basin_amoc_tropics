# <mark style="background-color: #E7C5FF"> <font color ='7C1931'> 1. Background </font> </mark>

Futher information and analysis can be found in a AMS 2024 Atmospheric and Oceanic Fluid Dynamics Conference poster linked [here](https://ams.confex.com/ams/24Fluid22Middle/meetingapp.cgi/Paper/443631).

Drifters package used is linked [here](https://github.com/cspencerjones/drifters).
Drifters were released for 60 years with initial positions of one particle/grid cell at $10^{\circ}S$ from $0^{\circ}E$ to $40^{\circ}E$. (40 lon $\times$ 32 depth = 1280 particles seeded). 


### <font color ='7C1931'> Purpose. </font>
These figures are used to better how changing background diapycnal diffusivity in a MOM6 idealized model simulation will affect the heat transport and water pathways of AMOC return flow water in the Tropical Atlantic. 

### <font color ='7C1931'> Drifters are divided into subsets. </font>
(drifter #, time drifter crosses chosen latitude) $\rightarrow$ drifter depth, longitude, Sv, theta = B $\subset$ 

Trajectories of drifters that reach 20$^{\circ}$N $\subset$ 

Trajectories of drifters that start at depths < 1000 = A $\subset$ 

Trajectories of all drifters.

Subsets A & B for [$\S 2$](#2) and [$\S 3$](#3) are created in the following notebooks.

|KD value | Notebook|
|:--------- |:-------- |
|5E-6 | [7-5_60yr_make_full_traj_20N_SB](./7-5_60yr_make_full_traj_20N_SB.ipynb)|
|1E-5 | [7-5_60yr_make_full_traj_20N_KD1E-5](./7-5_60yr_make_full_traj_20N_KD1E-5.ipynb)|
|0    | [7-10_60yr_make_full_traj_20N_KD0](./7-10_60yr_make_full_traj_20N_KD0.ipynb)|

### <font color='7C1931'> Regions. </font>
This notebook will reference regions in the depth-longitude plane. The goal of this is to differentiate between a Western Boundary Current region (R1), an interior region (R3), and a deep region (R2).

## <font color = 'F907035'> 2d. i) KD = 5E-6 </font>
<a name='2ci'></a>

```python
# get minimum time each particle crosses over the specified latitude. make dataset of that and collect Sv
chosen_lat_array = [-10.5,-0.5,5.5,10.5,20.5]
regions_and_lats = np.zeros((3,5))

for k in range(0,5):
    moving_n_at_t5 = above1000m_sb.where((above1000m_sb.vvel > 0) & (above1000m_sb.time ==5)).dropna('i',how='all')
    ivals = moving_n_at_t5.dropna('time',how='all').i.values
    moving_n_sincet5 = above1000m_sb.where(above1000m_sb.i.isin(ivals)).dropna('i',how='all')
    moving_n = moving_n_sincet5.where(moving_n_sincet5.vvel>0) #.dropna('i',how='all')
    moving_n_chosen_lat = moving_n.where(moving_n.lat >= chosen_lat_array[k]).dropna('i',how='all')
    
    predel_min_time_ind = moving_n_chosen_lat.drifter_num.argmin(dim='time').dropna('i').to_numpy()
    predel_i = moving_n_chosen_lat.drifter_num.argmin(dim='time').i.values#.dropna('i')
    predel_i_nans_ind = np.argwhere(np.isnan(predel_i))
    dn_times_del = np.delete(predel_min_time_ind,predel_i_nans_ind).astype(int)
    is_inds = np.argwhere(~np.isnan(moving_n_chosen_lat.i.values))
    
    mintimeind = moving_n_chosen_lat.lat.argmin(dim='time').values

    lon_list,lat_list,depth_list,drifter_num_list,vvel_list,depth_gridded_list,lon_gridded_list,Sv_list,theta_list = [],[],[],[],[],[],[],[],[]
    for i in range(0,len(mintimeind)):
        lon_list.append(moving_n_chosen_lat['lon'][mintimeind[i],i])
        lat_list.append(moving_n_chosen_lat['lat'][mintimeind[i],i])
        depth_list.append(moving_n_chosen_lat['depth'][mintimeind[i],i])
        drifter_num_list.append(moving_n_chosen_lat['drifter_num'][mintimeind[i],i])
        vvel_list.append(moving_n_chosen_lat['vvel'][mintimeind[i],i])
        depth_gridded_list.append(moving_n_chosen_lat['depth_gridded'][mintimeind[i],i])
        lon_gridded_list.append(moving_n_chosen_lat['lon_gridded'][mintimeind[i],i])
        Sv_list.append(moving_n_chosen_lat['Sv'][mintimeind[i],i])
        theta_list.append(moving_n_chosen_lat['theta'][mintimeind[i],i])
    dataset_names = [1,2,3,4,5,6,7,8,9]
    list_names = [lon_list,lat_list,depth_list,drifter_num_list,vvel_list,depth_gridded_list,lon_gridded_list,Sv_list,theta_list]
    datavar_names = ['lon','lat','depth','drifter_num','vvel','depth_gridded','lon_gridded','Sv','theta']
    for i in range(0,9):
        dataset_names[i] = xr.concat(list_names[i],dim='i').to_dataset(name=datavar_names[i])
    mintime_ds = xr.merge([dataset_names[0],dataset_names[1],dataset_names[2],dataset_names[3],
          dataset_names[4],dataset_names[5],dataset_names[6],dataset_names[7],dataset_names[8]])
    
    ### define regions
    region1 = mintime_ds.where((mintime_ds.lon_gridded <= 10) & (mintime_ds.depth_gridded < 400)).dropna('i',how='all')
    region2 = mintime_ds.where((mintime_ds.depth_gridded >= 400) & (mintime_ds.depth_gridded < 1000)).dropna('i',how='all')
    region3 = mintime_ds.where((mintime_ds.lon_gridded > 10) & (mintime_ds.depth_gridded < 400)).dropna('i',how='all')
    regions_and_lats[0,k] = region1.Sv.sum()
    regions_and_lats[1,k] = region2.Sv.sum()
    regions_and_lats[2,k] = region3.Sv.sum()
    
    total_volume_above_1000m = mintime_ds.Sv.sum() #where(mintime_ds.depth_gridded < 1000).Sv.sum()
```

```python
fig,ax=plt.subplots(nrows=1,ncols=1,constrained_layout=True)
latlabels = '-10.5N','-0.5N','5.5N','10.5N','20.5N'
barlabels = ['Region 1','Region 2','Region 3']

# colorblind colors
colors = ['#44BB99','#DDDDDD','#99DDFF']
a=0
for i in range(0,4):
    ax.bar(x=(0+a,0.8+a,1.6+a),height=regions_and_lats[:,i],color=colors)
    a=a+4

ax.bar(x=(0+a,0.8+a,1.6+a),height=regions_and_lats[:,i],color=colors,label=barlabels)
ax.legend()
ax.set_xticks(np.arange(0,20,4)+0.75,labels=latlabels)
ax.set_title('Volume Transport by Region for Northward Particles \n KD=5E-6 ')
ax.set_ylabel('Sv')
ax.set_xlabel('Latitude')
ax.set_ylim(0,12)
```
